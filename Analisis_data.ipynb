```python
import pandas as pd
import numpy as np

jumlah_data = 1000  

data = {
    "ID": range(1, jumlah_data + 1),
    "Usia": np.random.randint(18, 65, jumlah_data),
    "Jenis_Kelamin": np.random.choice(["Laki-laki", "Perempuan"], jumlah_data),
    "Pendapatan_Bulanan": np.random.randint(2, 50, jumlah_data),
    "Status_Pekerjaan": np.random.choice(["Pekerja", "Pelajar", "Tidak Bekerja"], jumlah_data),
    "Skor_Kredit": np.random.randint(300, 850, jumlah_data),
    "Jumlah_Transaksi": np.random.randint(0, 100, jumlah_data),
    "Kategori_Belanja": np.random.choice(["Makanan", "Fashion", "Elektronik", "Energi"], jumlah_data),
    "Kepemilikan_Properti": np.random.choice(["Ya", "Tidak"], jumlah_data),
    "Label_Klasifikasi": np.random.choice(["Layak", "Tidak Layak"], jumlah_data)
}

df = pd.DataFrame(data)
print(df.info())
print(df.head())

print("\nJumlah Jenis Kelamin:")
print(df["Jenis_Kelamin"].value_counts())

print("\nJumlah Status Pekerjaan:")
print(df["Status_Pekerjaan"].value_counts())

print("\nJumlah Label Klasifikasi:")
print(df["Label_Klasifikasi"].value_counts())

df.to_csv("dataset_sederhana.csv")
```
